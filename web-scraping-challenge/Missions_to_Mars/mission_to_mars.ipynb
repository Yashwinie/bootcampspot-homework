# NASA Mars News

#### Scrape the [NASA Mars News Site](https://mars.nasa.gov/news/) and collect the latest News Title and Paragraph Text. Assign the text to variables that you can reference later.

```python
#URL of page to be scraped
url = 'https://mars.nasa.gov/news/'
```

```python
# Retrieve page with the requests module
response = requests.get(url)
```

```python
# Create BeautifulSoup object; parse with 'html.parser'
soup = bs(response.text, 'html.parser')
```

```python
# Examine the results, then determine element that contains sought info
print(soup.prettify())
```

```python
# results are returned as a list
results = soup.find_all('div', class_="slide")
print(results)
```

```python
# Finding and saving latest news and latest news paragraph text
latest_news = results[0]
 
latest_newstitle = latest_news.find('div', class_='content_title').text
latest_newsp = latest_news.find('div', class_='rollover_description_inner').text

print(latest_newstitle, latest_newsp)
```

* Visit the Mars Weather twitter account [here](https://twitter.com/marswxreport?lang=en) and scrape the latest Mars weather tweet from the page. Save the tweet text for the weather report as a variable called `mars_weather`.

```python
# URL of page to be scraped
browser = webdriver.Chrome()
marsweather_base_url = 'https://twitter.com'
marsweather_distinct = '/marswxreport?lang=en'
mars_weather_url = marsweather_base_url + marsweather_distinct
```

```python
browser.get(mars_weather_url)
time.sleep(1)
```

```python
# Narrowing down to tweets by using Inspect, Beautiful Soup, and Selenium
body = browser.find_element_by_tag_name('body')
```

```python
for x in range(5):
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.2)
```

```python
tweets = browser.find_elements_by_class_name('tweet-text')
```

# Mars Facts

* Visit the Mars Facts webpage [here](https://space-facts.com/mars/) and use Pandas to scrape the table containing facts about the planet including Diameter, Mass, etc.

* Use Pandas to convert the data to a HTML table string.

```python
# URL of page to be scraped
marsfacts_base_url = 'https://space-facts.com'
marsfacts_distinct = '/mars/'
marsfacts_url = marsfacts_base_url + marsfacts_distinct
```

```python
# Retrieve page with the requests module
response = requests.get(marsfacts_url)
```

```python
# Create BeautifulSoup object; parse with 'html.parser'
soup = bs(response.text, 'html.parser')
```

```python
# Examine the results, then determine element that contains sought info
print(soup.prettify())
```

```python
# scrape the table containing facts about the planet by finding table tags
results = soup.find_all('table', class_='tablepress-id-p-mars')
marsfacts_df = pd.read_html(str(results))
print(marsfacts_df)
```

```python
#using pandas to convert table to HTML string
marsfacts_html = marsfacts_df.to_html()
```
